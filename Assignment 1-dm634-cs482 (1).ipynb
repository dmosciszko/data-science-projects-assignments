2. a- Signal occurs when at least 1 switch is closed 
$$P(SIGNAL) = P(S_1 \cup S_2 \cup S_3)$$
$S_i$ represents a closed switch and $\overline{S_i}$ an open switch.
$$P(S_i) = p$$
$$P(\overline{S_i}) = 1-p$$
$$P(S_1 \cup S_2 \cup S_3) = 1 - P(\overline{S_1} \cap \overline{S_2} \cap \overline{S_3}) $$
$$P(S_1 \cup S_2 \cup S_3) = 1-(1-p)(1-p)(1-p)$$
$$=1-(1-p)^3$$
$$P(SIGNAL)=3p-3p^2+p^3$$

  b- $$P(\overline{S_i} | SIGNAL) = \frac{P(SIGNAL | \overline{S_i}) * P(\overline{S_i})}{P(SIGNAL)} $$
  $$ = \frac{P(\overline{S_1} \cap S_2 \cap S_3) + P(\overline{S_1} \cap \overline{S_2} \cap S_3) + P(\overline{S_1} \cap S_2 \cap \overline{S_3})}{P(S_1 \cup S_2 \cup S_3)}$$
  $$ =\frac{(2p - p^2)}{3p -3p^2 + p^3}$$

```python
import random as rn
import matplotlib.pyplot as plt

prob_list = []
num_draws = []

for n in range(10, 1001):  #num of draws
  count = 0

  for i in range(n):
      x = rn.randrange(1, 11) #random int from 1 to 10
      if x== 2:
         count = count + 1
  prob = count / n
  prob_list.append(prob)  #plot
  num_draws.append(n) #plot

plt.plot(num_draws, prob_list)
plt.xlabel("Number of Draws")
plt.ylabel("Probability of 2")
plt.show()



```
